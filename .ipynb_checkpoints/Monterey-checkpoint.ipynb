# The Curve is Flat but Monterey Tightens Restrictions

### Yesterday, Monterey County Health commisioner issued [an order to wear facemask](https://www.co.monterey.ca.us/home/showdocument?id=90136) when outside performing essential activities. This comes at a time when it seems like the situation is getting better.

### First point is that there has been 4 deaths in the county and 185 confirmed cases. What does that mean for the death rate?

```python
dr = 4 / 185
print("Death Rate: 4/185 =", dr)
print("Survivability: (1 - death rate) * 100 =", 100 * (1 - dr))
```

### Granted that not all 185 have fully recovered yet, so there may be some more deaths. 39 have fully recovered, and according to the [daily situation report](https://www.co.monterey.ca.us/Home/ShowDocument?id=88023) there have been 30 hospitalizations. Where are the rest of the recoveries?

### Monterey reached 43 cases (39 recovered, 4 deaths) on the 30th of March. This virus has an average hospital stay of 7 days, and a [maximum stay of 24.](https://www.express.co.uk/news/uk/1258051/Coronavirus-patients-UK-hospitals-how-long-stay-in-hospital-covid-19) Which means that deaths are being recorded the day they happen while recoveries are sporadically being recorded and sometimes not at all. Most people recover at home and do not report it back.

### I think we could reasonably say that those who never went to the hospital are expected to recover. That would mean 4 deaths out of 155 cases

```python
dr = 4/155
print("Death Rate: 4/155 =", dr)
print("Survivability: (1 - death rate) * 100 =", 100 * (1 - dr))
```

### That means that right now you have about a 97% chance of recovery. Higher if those hospitalizations are recovered, lower if more than one of them dies while no new cases are added.

### Let's look at the graphs

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
dates = np.array(['2020-03-17', '2020-03-18'], dtype='datetime64')
for i in range(39):
    dates = np.append(dates, dates[-1] + 1)
```

```python
cases = np.array((2, 0, 0, 3, 2, 4, 3, 6, 4, 0, 3, 5, 4, 6, 6, 5, 4, 5, 0, 1, 1, 5, 2, 8, 3, 5, 0, 13, 8, 11, 17, 0, 5, 7, 2, 4, 10, 5, 7, 7, 2))
case_cumsum = np.cumsum(cases)
```

```python
print("len(dates) =", len(dates), "\nlen(cases) =", len(cases), "\nlen(case_cumsum) =", len(cases))
```

### Here is the graph from the situation report:

```python
fig, ax = plt.subplots(figsize=(13, 13))
plt.xticks(fontsize=14, rotation=60)
plt.yticks(fontsize=14)
ax.plot(dates, case_cumsum)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Cases', fontsize=14)
plt.show()
```

### Looks pretty bad right?

### That is NOT the curve that we are trying to flatten. That is the the cumulative sum of cases. In other words it will always go up. It cannot go down. It will not flatten until the virus is erradivated from the earth, and then it will stay at the highest point on to infinity.

### Let's look at the curve we are trying to flatten:

```python
fig, ax = plt.subplots(figsize=(13, 13))
plt.xticks(fontsize=14, rotation=60)
plt.yticks(fontsize=14)
ax.plot(dates, cases)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Cases', fontsize=14)
plt.show()
```

### This is the curve we are trying to flatten in order to not overwhelm the hospitals. Looks like our county peaked around 18 in a day. We have 30 total hospitalizations through the entire epidemic. We don't know how many of those are still in the hospital, but with an average of 7 days per visit, we can assume most are not in the hospital anymore. With 4 hospitals, did we ever come close to overwhelming the system?

### Another important note is that this is according to "Date Reported". Other studies are starting to show that the report date is very sporadic compared to Date Of Illnes Onset (DOIO) and cases pop up later making these graphs look like they are still growing when they actually grew more earlier, and are much less now.

### But let's assume all the data is perfect. If we add a trend line:

```python
xnew = np.array(range(len(cases)))
polynomial_coeff = np.polyfit(xnew, cases, 1)
ynew = np.poly1d(polynomial_coeff)
fig, ax = plt.subplots(figsize=(13, 13))
plt.xticks(fontsize=14, rotation=60)
plt.yticks(fontsize=14)
ax.scatter(dates, cases)
ax.plot(dates, ynew(xnew), c='r')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Cases', fontsize=14)
plt.show()
```

### Looks like we have a positive trend, but what if we make it a polynomial trend line?

```python
polynomial_coeff = np.polyfit(xnew, cases, 2)
ynew = np.poly1d(polynomial_coeff)
fig, ax = plt.subplots(figsize=(13, 13))
plt.xticks(fontsize=14, rotation=60)
plt.yticks(fontsize=14)
ax.scatter(dates, cases)
ax.plot(dates, ynew(xnew), c='r')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Cases', fontsize=14)
plt.show()
```

### Even if that doesn't look like a downward trend to you, does it look like we should be increasing restrictions?

### The director said "[Because it's working, I'm intending on extending the order](https://www.ksbw.com/article/monterey-county-intends-to-extend-shelter-in-place-order/32275040). By that logic, this will never end. He is not only extending it, he is increasing restrictions. By the end of the year, we will be confined to our bathrooms.

* April 24, [Closed hotels](https://www.co.monterey.ca.us/home/showdocument?id=90081) except for certain situations.
* April 26, [Closed more beaches](https://www.montereyherald.com/2020/04/26/officials-close-lovers-point-in-the-name-of-social-distancing/)
* April 28, [Required Face Coverings](https://www.co.monterey.ca.us/Home/Components/News/News/7283/1336?backlist=%2fgovernment%2fdepartments-a-h%2fhealth) outside.

### Maybe he is looking at something else. Some articles point out the [doubling rate](https://www.ksbw.com/article/monterey-county-intends-to-extend-shelter-in-place-order/32275040). 
* The curve for Monterey County has been relatively flat compared to the rest of the state. According to an LA Times infographic, coronavirus cases are doubling in the county every 16 and a half days.

### What is the doubling rate in Monterey County?

```python
xvals = np.linspace(0, len(cases), 10000)
x_dates = np.linspace(pd.Timestamp(dates[0]).value, pd.Timestamp(dates[-1]).value, 10000)
x_dates = pd.to_datetime(x_dates)
yinterp = np.interp(xvals, xnew, case_cumsum)
doubling_rate = np.ones(len(yinterp))
for i, j in enumerate(yinterp):
    half = yinterp[i] / 2
    if half <= 2:
        doubling_rate[i] = np.nan
    else:
        x = np.datetime64(x_dates[i]) - np.datetime64(x_dates[yinterp== yinterp[np.abs(yinterp - half).argmin()]][0])
        days = x.astype('timedelta64[D]')
        days_as_int = days / np.timedelta64(1, 'D')
        doubling_rate[i] = days_as_int
fig, ax = plt.subplots(figsize=(13, 13))
plt.xticks(fontsize=14, rotation=60)
plt.yticks(fontsize=14)
ax.plot(x_dates, doubling_rate)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Doubling Rate', fontsize=14)
plt.show()
```

### Doubling time is the time it takes takes to double the number of cases. It is a good indicator of how fast it is spreading. It does not work as well when we do not know Date of Illness Onset, because if some cases are not reported every day, and then all of those are reported on one day, it artificially decreases the doubling time.

### A doubling time of 3 days is what we wanted to avoid. That would mean there is runaway spread. Monterey County is increasing every day. Currently at 14 days. Why would we INCREASE restrictions?
